```python
import pandas as pd
df = pd.read_csv('/content/world_population.csv')
```

```python
print('Columns in the DataFrame:')
print(df.columns.tolist())
```

```python
df.isnull().mean()
```

```python
df.drop_duplicates()
```

```python
population_columns = [
    '1970 Population', '1980 Population', '1990 Population', '2000 Population',
    '2010 Population', '2015 Population', '2020 Population', '2022 Population'
]
display(df[population_columns].describe())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(df['Continent'], df['Area (km²)'])
plt.xlabel('Continent')
plt.ylabel('Area (km²)')
plt.title('area  Countries')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['Continent'], df['2022 Population'])
plt.xlabel('Continent')
plt.ylabel('2022 Population')
plt.title('Population of Countries')
plt.show()
```

```python
continent_population_percentage = df.groupby('Continent')['World Population Percentage'].sum()

plt.figure(figsize=(8, 8))
plt.pie(continent_population_percentage, labels=continent_population_percentage.index, autopct='%1.1f%%', startangle=90)
plt.title('World Population Percentage by Continent')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(90,6))
plt.plot(df['Country/Territory'], df['Growth Rate'])
plt.xlabel('Country/Territor')
plt.ylabel('Growth Rat')
plt.title('World Population Percentage Over Time')
plt.show()
```
