```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a sample dataset
np.random.seed(42)
data = {
    'Feature_A': np.random.rand(100) * 100,
    'Feature_B': np.random.rand(100) * 50 + 20,
    'Feature_C': np.random.randint(0, 5, 100),
    'Target': np.random.rand(100) * 100 + (np.random.rand(100) * 10)
}
df = pd.DataFrame(data)

df['Feature_C'] = df['Feature_C'].astype('category')

# Display the first few rows of the dataset
display(df.head())
```

```python
# Create a scatter plot of Feature_A vs Target
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Feature_A', y='Target', data=df, hue='Feature_C', palette='viridis')
plt.title('Scatter Plot of Feature A vs Target')
plt.xlabel('Feature A')
plt.ylabel('Target')
plt.grid(True)
plt.show()
```

```python
# Create a histogram of Feature_B
plt.figure(figsize=(10, 6))
sns.histplot(df['Feature_B'], kde=True, bins=15, color='skyblue')
plt.title('Distribution of Feature B')
plt.xlabel('Feature B')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create a box plot of Target by Feature_C
plt.figure(figsize=(10, 6))
sns.boxplot(x='Feature_C', y='Target', data=df, hue='Feature_C', palette='pastel', legend=False)
plt.title('Box Plot of Target by Feature C')
plt.xlabel('Feature C')
plt.ylabel('Target')
plt.grid(axis='y')
plt.show()
```
